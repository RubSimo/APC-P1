# Apartat B: Predicció


Ara el nostre objectiu és la predicció d'un atribut. L'edat és un valor que es pot identificar fàcilment, es pot aconseguir per informació bàsica que es demana al pacient (Nom i cognoms, edat, sexe, ect.). La pressió en sang és una prova senzilla, que no requereix molt de temps, per tant, és fàcil d'aconseguir. En canvi una prova de colesterol és més complexe, cal estar hores sense ingerir aliment seguit d'un anàlisi de sang. És un atribut que pot ser d'utilitat però la complexitat està en que hi ha molts factors que depenen del colesterol. És un valor molt variable en un rang bastant gran de valors. És difícil de predir. I per últim la freqüència cardíaca en repòs. Un valor que més o menys coincideixen amb diferents característiques de les persones, ja pot ser l'edat, sexe, pes... Són dades que en aproximadament 300 pacients poden tenir una certa coincidència. Tot i que no és un atribut especialment difícil d'aconseguir, és el que pensem que pot tenir més correlació amb els altres atributs i obtenir un baix error de predicció.

A continuació definim les funcions bàsiques per poder obtenir resultats en la predicció.


```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import sklearn

def regression(x, y):
    regr = LinearRegression()
    regr.fit(x, y)
    return regr

def mse(v1, v2):
    return ((v1 - v2)**2).mean()

def standarize(x_train):
    mean = x_train.mean(0)
    std = x_train.std(0)
    x_t = x_train - mean[None, :]
    x_t /= std[None, :]
    return x_t
```

En aquest cas farem la predicció amb cadascun dels atributs com entrada. Calcularem l'error quadràtic i es mostrarà la gràfica amb la recta de predicció. Veurem quin atribut es el més indicat pero obtenir bons resultats.

```python
data = dataset.values
x = data[:,[0,1,2]]
y = data[:,3]


x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size =0.3, random_state = 42)
    
    
for i in range(x_train.shape[1]):
    x_tr = x_train[:,[i]]
    x_te = x_test[:,[i]]
    regr = regression(x_tr, y_train)
    predicted = regr.predict(x_te)
    error = mse(y_test, predicted)
    r2 = r2_score(y_test, predicted)
    plt.figure()
    ax = plt.scatter(x[:,i], y)
    plt.plot(x_te, predicted, 'r')
    print("Error en atribut %d: %f" %(i, error))
    print("R2 score en atribut %d: %f" %(i, r2))

```

Les gràfiques indiquen quins atributs estàn més dispersos i quins no. Podem veure que la primera gràfica, de l'atribut 0 (edat) té una dispersió més petita que la resta. Amb el paràmetre random_state = 42 obtenim que el màxim de r2 score es per aquest atribut, és el més alt dels 3. La funció split divideix aleatòriament i com a màxim aconseguim un r2 score de 0.30. Aquest és un valor molt baix i podem arribar a la conclusió que no es un bon model per aquestes dades.


La gràfica que mostrem a continuació mostra com podem trobar sentit als resultats de la funció anterior. Veiem que la correlació entre l'edat i la frequencia caridaca es la més alta.

```python
correlacio = dataset.corr()

plt.figure()

ax = sns.heatmap(correlacio, annot=True, linewidths=.5)
```

Vistos tots els resultats, optem per normalitzar les dades. Les consequències de la falta de normalització poden ser:
1. Inexactitud de dades. 
2. Relentizació de processos. 
3. Ineficiècia d'operacions.
    
És un procediment fundamental dins del món laboral, amb que podem aconseguir optimitzar les nostres dades i per tanto, obtenir una millora en la predicció.


```python
data = dataset.values
#normalitzem tots els atributs
norm_data=[]

for i in range(data.shape[1]):
    norm_data.append(standarize(data[:,[i]]))

norm_data = np.array(norm_data)
norm_data = norm_data.reshape((norm_data.shape[0],norm_data.shape[1]))
norm_data = norm_data.transpose()
```

Una vegada normalitzats, tornem a fer la predicció per comparar resultats.

```python
x = norm_data[:,[0,1,2]]
y = norm_data[:,3]

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.3, random_state=42)
```

```python
for i in range(x_train.shape[1]):
    x_tr = x_train[:,[i]]
    x_te = x_test[:,[i]]
    regr = regression(x_tr, y_train)
    error = mse(y_test, regr.predict(x_te))
    error2 = sum( (y_test-regr.predict(x_te))**2)/(2*y_test.shape[0])
    r2 = r2_score(y_test, regr.predict(x_te))
    
    print('Error de comparació amb Descens de Gradient: %f' %(error2))
    print("Error en atribut %d: %f" %(i, error))
    print("R2 score en atribut %d: %f" %(i, r2))
```

La diferència de resultats no varía. La causa pot ser l'escassetat de dades que tenim.

Provem a filtrar els valors d'entrada, reduint exepcions.

```python
dat = dataset.values
dat = dat[(dat[:,1]<180) & (dat[:,2]<400)] # pressió en sang < 180 i colesterol < 400 (superar aquests numeros son casos extrems)

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(dat[:,:3], dat[:,-1], test_size = 0.3, random_state=42)

for i in range(x_train.shape[1]):
    x_tr = x_train[:,[i]]
    x_te = x_test[:,[i]]
    regr = regression(x_tr, y_train)
    error = mse(y_test, regr.predict(x_te))
    r2 = r2_score(y_test, regr.predict(x_te))
    
    print("Error en atribut %d filtrat: %f" %(i, error))
    print("R2 score en atribut %d filtrat: %f" %(i, r2))
```

Donat que tenim tan poques dades, si eliminem aquelles que son casos anormals, estem perdent dades i obtenim una predicció més imprecisa. Filtrar valors dels atributs queda descartat.


Per intentar millorar el nostre model, afegirem més atributs d'entrada per veure quin es el resultat.

```python
x = norm_data[:,[0,1,2]]
y = norm_data[:,3]
x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size = 0.3, random_state=42)
comb = [[0,1],[0,2],[1,2]]    
for i in comb:
    x_tr = x_train[:,i]
    x_te = x_test[:,i]
    regr = regression(x_tr, y_train)
    error = mse(y_test, regr.predict(x_te))
    r2 = r2_score(y_test, regr.predict(x_te))
     
    print("Error en atributs %s: %f" %(i, error))
    print("R2 score en atributs %s: %f" %(i, r2))
```

Com hem vist que amb 3 dimensions d'entrada es la millor opció, aplicarem PCA per reduir les dimensions i comprovar si aconseguim millorar la predicció i eliminarem els valos nan.

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

d = StandardScaler().fit_transform(dataset) #normalitzem totes les dades
pca = PCA(n_components=2) 
new_x = pca.fit_transform(d[:,:3]) #reduim dades d'entrada a 2 dimensions
new_data = pd.concat([pd.DataFrame(data = new_x), pd.DataFrame(data = d[:,-1])], axis = 1) #afegim les x reduides i les y a una taula
new_data = new_data.dropna()
new_data = new_data.to_numpy()

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(new_data[0:,:2], new_data[:,-1], test_size = 0.3, random_state = 42)
regr = regression(x_train, y_train)
error = mse(y_test, regr.predict(x_test))
r2 = r2_score(y_test, regr.predict(x_test))

print("Error en atributs: %f" %(error))
print("R2 score en atributs: %f" %(r2))
```

# Apartat A: descenso del gradiente
"""
S'ha realitzat la classe Regressor i te la capacitat de realitazar
regressio lineal simple i multiple.
Tambe te funcions per pintar grafics.
"""

```python
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
class Regressor(object):
    def __init__(self, w0, listaW, listax, y, alpha):
        # Inicialitzem w0 i w1 (per ser ampliat amb altres w's)
        self.m=len(y)                             #
        self.num_variables=len(listaW)
        self.alpha=alpha*(1/len(y))
        self.listaW=listaW
        self.w0=w0
        self.listaX=listax
        self.y=y
        self.historial_coste=list()
        self.coste_modelo=0

        self.sumatorio_hipotesis=np.zeros(len(listaW)+1)
        self.historial_W0=list()
        self.historial_W1=list()
    'w0 +w1*x2+ w2*x2 ... wnxn'
    def hipotesis(self, listaX):
        sum=0.0
        #Realitzat de aquesta manera per poder concatenar variables i pesos
        for i in range(self.num_variables):
            sum += self.listaW[i]*listaX[i]
        return sum+self.w0
    'Funcio que agrupa en una llista el diferents valors de cada parametre'
    def agrupar(self,fila):
        lista_aux=np.zeros(self.num_variables)
        for col in range(self.num_variables):
            lista_temp=self.listaX[col]
            lista_aux[col]=lista_temp[fila]
        return lista_aux
    'Funcio per fer el sumatori dels costos'
    def sumatori_Hipotesis(self,coste_derivado,lista_aux):
        self.sumatorio_hipotesis[0]+=coste_derivado
        for col in range(self.num_variables):
            self.sumatorio_hipotesis[col+1]+=coste_derivado*lista_aux[col]
    '''Funcio que realitza la suma de costos sense derivar
    I la suma per cada pes multiplicat per la variable'''
    def coste(self):
        coste_sin_derivar=0.0   
        for i in range(self.m):
            #agrupar els valors de cada variables depenent de i
            lista_aux=self.agrupar(i)  
            #Tenir el cost derivat per despres canviar el valor dels pesos
            coste_derivado=self.hipotesis(lista_aux)-self.y[i]
            #Tenir el cost sence derivar per fer el error mitx
            coste_sin_derivar+=pow(coste_derivado,2)
            self.sumatori_Hipotesis(coste_derivado,lista_aux) 
        return coste_sin_derivar
    'Funcio per modificar el pesos'
    def modificar_pesos(self):
            for col in range(self.num_variables):
                self.listaW[col]-=self.alpha*self.sumatorio_hipotesis[col+1]
                self.sumatorio_hipotesis[col + 1]=0
            self.w0-=self.alpha*self.sumatorio_hipotesis[0]
            self.sumatorio_hipotesis[0]=0
    '''Fucio que realitza el entrenament
    Se anirà ajustant el pesos fins, intentant arribar a un mínim'''
    def train(self,max_ite):
        for i in range(max_ite):
            #La mitxa de error per cada iteració
            #servira per fer al gràfica'''
            self.historial_coste.append(self.coste()/(2*self.m))
            #afegir el valor dels pesos per despres poder fer gràfiques
            self.historial_W0.append(self.w0)
            self.historial_W1.append(self.listaW[0])
            #canviar el valor dels pesos
            self.modificar_pesos()
        #afegim el ultim registes del historial de costes, aquest valor en indica el cost del model  
        self.coste_modelo=self.historial_coste[-1]

    '''Funcio que ens servira per printar gráfiques
    per una llista qualsevol de input, es cercara el ouput amb 
    el pasos que hagem consguit en el nostre model.
    De aquesta manera acabarem pintant la linea de regresio.
    '''
    def print_grafic(self,tipus):
        listaX=list()
        size=100
        y = np.zeros(size)
        listaX.append(np.linspace(np.min(self.listaX[0]), np.max(self.listaX[0]), size))

        for i in range(1,self.num_variables,):
            listaX.append(listaX[0] ** i+1)
        for it in range(size):
            sum = 0.0
            for i in range(self.num_variables):
                temp=listaX[i]
                sum += self.listaW[i]*temp[it]
            y[it]=sum+self.w0
        #grafic 2d
        if(tipus):
            plt.figure(figsize=(8, 6))
            plt.title("Regressió",fontdict={'family': 'serif','color' : 'darkblue','weight': 'bold','size': 18})
            plt.plot(self.listaX[0], self.y, 'o', markersize=2)
            plt.plot(listaX[0], y)
            plt.show()
        #grafic 3d
        else:
            xplot,yplot=np.meshgrid(listaX[0] ,listaX[1])
            zplot=self.listaW[0] * xplot + self.listaW[2]  * yplot + self.w0

            plt3d = plt.figure('Coeficiente prismatico -- Relacio longitud desplacament 3D', dpi=100.0).gca(projection='3d')
            plt3d.plot_surface(xplot,yplot,zplot, color='red')
            plt3d.scatter(self.listaX[0], self.listaX[1], self.y, c='g', marker='o')

            plt.show()

   
    '''Aquesta funció es unicament per visualitza una grafica 3d
    amb el valor de costos per tota convinació de pesos, i de aquesta 
    manera peder veure de una manera gráfica com es realitza el descens.
    '''
    def print_HistorialW(self):
        size=100
        w0=np.linspace(-7, 6, size)
        w1=np.linspace(-7, 6, size)
        zplot=np.zeros((size,size))
        lista_temp = self.listaX[0]
        for i in range(size):
            for j in range(size):
                coste = 0.0
                for fila in range(self.m):
                    sum = 0.0
                    sum += w1[i] * lista_temp[fila] +w0[j]
                    coste += pow(sum - self.y[fila],2)
                zplot[i][j]=(coste/(2*self.m))
        w0_plot, w1_plot = np.meshgrid(w0 , w1)
        self.train(1000)


        plt3d = plt.figure('Coste', dpi=100.0).gca(projection='3d')
        plt3d.contour3D(w0_plot,w1_plot,zplot, 50 ,cmap=cm.coolwarm, antialiased=False)       
        plt3d.view_init(45, 0) 
        plt3d.scatter(self.historial_W1, self.historial_W0, self.historial_coste, c='g', marker='o',linewidth=2)
        plt3d.scatter(self.historial_W1[0], self.historial_W0[0], self.historial_coste[0], c='black', marker='o',linewidth=10)
        plt3d.scatter(self.historial_W1[-1], self.historial_W0[-1], self.historial_coste[-1], c='black', marker='o',linewidth=10)

        plt.title("Descens del Gradient",fontdict={'family': 'serif','color' : 'darkblue','weight': 'bold','size': 18})
        plt.xlabel("W0", size = 16,)
        plt.ylabel("W1", size = 16)
   
        plt.show()
    'Funcio per observa el conportament del cost per el numero de iteracion realitzades'
    def print_coste(self):
        plt.figure(figsize=(8, 6))
        plt.title("Cost",fontdict={'family': 'serif','color' : 'darkblue','weight': 'bold','size': 18})

        plt.plot(np.arange(len(self.historial_coste)), self.historial_coste)
        plt.xlabel("Iteracions", size = 16,)
        plt.ylabel("Cost", size = 16)

        plt.show()
```

Test coeficiente de aprenentatge
Aquest valor definira com de gran es el canvi que es realitzara
els nostres pesos.
Segons la teoria aquest valor ha de anar entre [0,1], pero 
per observar els efectes de aquest valor, s'ha sobrepasat aquesta franja.
S'ha definit amb tres valors[2.5,0.01,0.0000001]
- Si el valor es massa gran, els canvis als pesos seran massa drastic anan de una punta a un altre 
sense poder aribar a convergir i elevan el cost del model.
- Per un valor normal: per el nostre cas el valor no cal que sigui molt petit, el motiu es perque la 
    complexitat de les dades no es elevadad i en el cas del exemple es una regresio lineal, per tan 
    en pocs pasos s'arriba al minim.
- Si el valor es massa petit: la convergensia es molt lenta i caldria augmentar molt 
    el numero de iteracion perque arribi a convergir.


```python
import random as rnd


y=norm_data[:,3]
x=norm_data[:,0]
w0=rnd.uniform(np.min(y),np.max(y))
coeficiente=[2.5,0.01,0.0000001]

print("Test cambiar alpha")
for i in range(0,3):
    regresor = Regressor(w0, [rnd.uniform(0,1)], [x], y, coeficiente[i]);
    regresor.train(500)
    regresor.print_grafic(1)
    regresor.print_coste()
    print("Cost del model: ",regresor.coste_modelo)
```

Test numero de iteracions
Es pot observar, com a més iteracions el cost va disminuin i cada cop ens acostem
més al minim, pero això dependra molt del valor que tinguem de coeficiente de aprenentatge.
Per tant a de haver un equilibri entre el numero de iteración i el coeficient de aprenentage.

```python
print("Test cambiar numero de ejecuciones")
iteraciones=[100,1000,2000,5000,10000,20000]
for i in range(len(iteraciones)):
    regresor = Regressor(w0, [rnd.uniform(0,1)], [x], y, 0.0002);
    regresor.train(iteraciones[i])
    regresor.print_grafic(1)
    regresor.print_coste()
    print("Cost del model:",regresor.coste_modelo)
```

La clase Regressor s'ha fet amb la capasitat de realitzar regresio multivarible.
En aquesta prova en contes de afegir diferents variables, s'ha augmentat el grau.
D'aquesta manera es dota del model pero poder obter un resultat més "flexible"
i de aquesta manera el model es pot apabtar millor a les dades i aconseguir baixar el cost.

Com observacion, el cost es conseguix diminuir en comparaciño del regresor lineal,
pero amb millores molt petites degut a que les dades estan molt dispersas, i ha poca correlació entre elles
com a consecuencia, una model lineal simple es suficient perque casi no val la pena
dotar al model mes flexiblitat.

```python
print("Test grado del desenso")
X=list()
W=list()
for i in range(4):
    print("Grau: ",i+1)
    X.append(x**(i+1))
    W.append(rnd.uniform(0,1))
    regresor = Regressor(w0, W, X, y, 0.01);
    regresor.train(1000)
    regresor.print_grafic(1)
    regresor.print_coste()
    print("Cost del model: ",regresor.coste_modelo)
```
